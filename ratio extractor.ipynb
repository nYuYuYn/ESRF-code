The purpose of the script is to normalize or correct the intensity in SAXS data. To do so, the script will try to:

#### 1. Extract the data of "SAXSdio" & "mondio" and calculate the ratio (SAXSdio/mondio):
#### 2. Save the ratio as a csv file.

```python
import numpy as np
import h5py
import pandas as pd
```

```python
file_path='C:/Users/opid31/Desktop/h5 file/01_lithiation_0001.h5'

dfs=[]

with h5py.File(file_path, 'r') as f:
    
    a=list(f.keys())
    a.remove('2.1')
    for i in a:
        #print(i)
        G = f.get('%s'%i)
        
        path = '/%s/instrument/' %i
        
        G_saxsdio = G.get(path + 'scaled_saxsdio')
        #print(G_saxsdio)
        G_saxsdio_data = np.array(G_saxsdio.get('data'))
        #print(G_saxsdio_data)
        
        G_mondio = G.get(path + 'scaled_mondio')
        #print(G_mondio)
        G_mondio_data = np.array(G_mondio.get('data'))
        #print(G_mondio_data)

        G_ratio = np.divide(G_saxsdio_data, G_mondio_data)
        #print(G_ratio)
        
        j=float(i)

        df = pd.DataFrame(G_ratio, columns=['%.0f'%j])
    
        dfs.append(df)
    
    ratiodfs = pd.concat(dfs, axis=1)
    ratiodfs.to_csv('ratio_01_lithiation.csv', index=False)
    
    
        
```
